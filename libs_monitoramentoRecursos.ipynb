EXECUTANDO O PROCESSO DE COLETA 

```python
import pathlib as path
import psutil as systemMonitor
import pandas as pd

qtdTempoMonitoramentoDados = int(input("Insira o tempo desejo para monitoramento (Em segundos)"))

dadosMonitoramento = []

def monitorarMemoriaDoComputador(qtdTempoMonitoramentoDados):
    # print("Tempo de monitoramento: %d", qtdTempoMonitoramentoDados)

    percentualCpu = systemMonitor.cpu_percent(interval= 1)
    memoriaTotal = systemMonitor.virtual_memory().total
    memoriaUsada = systemMonitor.virtual_memory().used
    memoriaDisponivel = systemMonitor.virtual_memory().free
    memoriaUtilizadaPercentual = systemMonitor.virtual_memory().percent
    bytesEnviados  = systemMonitor.net_io_counters().bytes_sent
    bytesRecebidos = systemMonitor.net_io_counters().bytes_recv

    for i in range(qtdTempoMonitoramentoDados):
        dadosMonitoramento.append(percentualCpu)
        dadosMonitoramento.extend(memoriaTotal)
        dadosMonitoramento.extend(memoriaUsada)
        dadosMonitoramento.extend(memoriaDisponivel)
        dadosMonitoramento.extend(memoriaUtilizadaPercentual)
        dadosMonitoramento.extend(bytesEnviados)
        dadosMonitoramento.extend(bytesRecebidos)


    ExibirMonitoramento(percentualCpu, memoriaTotal, memoriaUsada, memoriaDisponivel, 
    memoriaUtilizadaPercentual, bytesEnviados, bytesRecebidos)




def criarExcelParaMonitoramentoDoComputador():
    df = pd.DataFrame(dadosMonitoramento, columns=['UsoCPU'])
    df.to_excel('minha_lista.xlsx', index=False)



def ExibirMonitoramento(cpu, memTotal, memUsada, memDisp, memPercent, bytesEnv, bytesRec):
    print(cpu)
    print(memTotal)
    print(memUsada)
    print(memDisp)
    print(memPercent)
    print(bytesEnv)
    print(bytesRec)

monitorarMemoriaDoComputador(qtdTempoMonitoramentoDados)

print(dadosMonitoramento)

criarExcelParaMonitoramentoDoComputador()

```
