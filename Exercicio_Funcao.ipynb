```python
import numpy as np
import matplotlib.pyplot as plt

def plotarGraficos(x, y):
# Criação do gráfico de linhas
    plt.figure(figsize=(12, 8))
    plt.plot(x, y, 'b-', label='x', linewidth=2.5)


    plt.xlabel('x', fontsize=12)
    plt.ylabel('f(x)', fontsize=12)
    plt.title('Gráficos das funções no intervalo [0, 10] (100 pontos)', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)
    plt.xlim(0, 10)
    plt.tight_layout()
    plt.show()

def f_x_ao_quadrado(x):
    return x**2

def f_x_raiz(x):
    return np.sqrt(x)

def f_sen_cos(x):
    return np.sin(x) + np.cos(x)

def f_sen_potencia(x):
    sin_x = np.sin(x)
    return np.power(np.abs(sin_x), 2.7) * np.sign(sin_x) - 0.1 * x


x = np.linspace(0, 10, 100)

y1 = f_x_ao_quadrado(x)
y2 = f_x_raiz(x)
y3 = f_sen_cos(x)
y4 = f_sen_potencia(x)

plotarGraficos(x, y1)
plotarGraficos(x, y2)
plotarGraficos(x, y3)
plotarGraficos(x,y4)



    
```
