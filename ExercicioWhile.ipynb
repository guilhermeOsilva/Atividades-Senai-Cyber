4) Utilizando o programa de carregamento de imagem fornecido pelo
professor, implemente um programa em Python que tira a média de 4
pixels vizinhos e substitui este valor médio nos mesmo pixels que foram
utilizados para gerar a média. A figura abaixo ilustra o procedimento.

```python
import cv2
import numpy as np

imagem=cv2.imread("images.jpg",cv2.IMREAD_GRAYSCALE)
altura,largura=imagem.shape

nova_altura = altura // 2
nova_largura = largura // 2
imagem2=np.zeros((nova_altura,nova_largura), dtype=np.uint8)

i = 0
while i < nova_altura:
    j = 0
    while j < nova_largura:
        y = i * 2
        x = j * 2
        
        regiao=imagem[y:y+2,x:x+2]
        media=int(np.sum(regiao)/4)  
        imagem2[i,j]=media             
        
        j += 1
    i += 1

cv2.imshow("Original",imagem)
cv2.imshow("Média 2x2 (4 pixels)",imagem2)
cv2.imwrite("questao4.png",imagem2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

5) Modifique o programa anterior implementando grupos de 9, 25 e 36
pixels e verifique o resultado.

```python
import cv2
import numpy as np

imagem=cv2.imread("images.jpg",cv2.IMREAD_GRAYSCALE)
altura,largura=imagem.shape

grupos = [(3,9), (5,25), (6,36)]

for n, total_pixels in grupos:
    print(f"\nProcessando {n}x{n} = {total_pixels} pixels")
    
    nova_altura = altura // n
    nova_largura = largura // n
    imagem2=np.zeros((nova_altura,nova_largura), dtype=np.uint8)

    i = 0
    while i < nova_altura:
        j = 0
        while j < nova_largura:
            y = i * n
            x = j * n
            
            regiao=imagem[y:y+n,x:x+n]
            media=int(np.sum(regiao)/total_pixels)
            imagem2[i,j]=media
            
            j += 1
        i += 1
    
    nome_arq = f"gato_{n}x{n}.png"
    cv2.imwrite(nome_arq, imagem2)
    print(f"Salvou {nome_arq}")

cv2.waitKey(0)
cv2.destroyAllWindows()
```

6) Modifique os programas 4 e 5 utilizando imagens coloridas. Você
deverá calcular 3 médias, uma para o verde, uma para o vermelho e
uma para o azul.

```python
import cv2
import numpy as np

imagem = cv2.imread("images.jpg",cv2.IMREAD_COLOR)
altura, largura, t = imagem.shape

nova_altura = altura // 2
nova_largura = largura // 2
imagem2 = np.zeros((nova_altura,nova_largura,3), dtype=np.uint8)

i = 0
while i < nova_altura:
    j = 0
    while j < nova_largura:
        y = i * 2
        x = j * 2
        
        regiao=imagem[y:y+2,x:x+2]           
        media=np.mean(regiao,axis=(0,1))
        imagem2[i,j]=media.astype(np.uint8)
        
        j += 1
    i += 1

cv2.imshow("Original Color",imagem)
cv2.imshow("Média 2x2 Color (R/G/B)",imagem2)
cv2.imwrite("questao6_color.png",imagem2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
